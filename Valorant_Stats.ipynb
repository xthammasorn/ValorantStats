```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv("valorant_stats.csv")
df
```

```python
df = pd.read_csv("valorant_stats.csv")
spec_chars = [ ","]
for i in spec_chars:
  df['Damage'] = df['Damage'].str.replace(i,"")
  df['Damage'] = df['Damage'].astype(float)
df
```

```python
map = []
Agent = []
k = []
d = []
kd=[]
```

```python
#K/D
for i in df['Match']:
  K = df['Kill'][i-1]
  D = df['Death'][i-1]
  KD = df['Kill'][i-1] / df['Death'][i-1]
  k.append(K)
  d.append(D)
  kd.append(KD)
```



*   **K/D**



```python
k = k
d = d
match = df['Match']
```

```python
plt.plot(match,k , "r" , marker=".")
plt.plot(match,d , "b", marker=".")
data = {"size":10  ,"color":"white","backgroundcolor":"#808080"}
plt.xlabel("Match",data)
plt.ylabel("Kill or death",data)
tittleconfig = {"size":15 , "color":"#800080"}
plt.title("Kill Death",tittleconfig)
plt.legend(["Kill","Death"])
plt.show()
```




*   **Map**











```python
Haven = 0
Bind = 0
Fracture	= 0
Split = 0
Icebox = 0
Ascent = 0

Map = ["Haven" , "Bind","Fracture","Split","Icebox","Ascent"]

for j in range(len(df["Map"])):
    if df["Map"].loc[j] == "Haven" :
      Haven = Haven + 1
    if df["Map"].loc[j] == "Bind" :
      Bind = Bind + 1
    if df["Map"].loc[j] == "Fracture" :
      Fracture = Fracture + 1
    if df["Map"].loc[j] ==  "Split" :
      Split = Split + 1
    if df["Map"].loc[j] == "Icebox" :
      Icebox = Icebox + 1
    if df["Map"].loc[j] == "Ascent" :
      Ascent = Ascent + 1

Mapp = [Haven,Bind,Fracture,Split,Icebox,Ascent]
color = ["#DAA520","#808000","#2E8B57","#3CB371","#87CEEB","#B22222"] 
plt.pie(Mapp,labels = Map ,colors=color, autopct='%.1f%%',startangle= 45)
plt.title("Map")
plt.show()
print()
```

```python
exp = [0,0.2,0,0,0,0]
plt.pie(Mapp,labels = Map ,colors=color, autopct='%.1f%%',explode=exp , startangle= 45)
title_text = {"size":15}
plt.title("Map",title_text)
plt.show()
print()
```


*  **Agent**




```python
Astra = 0
Sova = 0
Jett	= 0
Reyna = 0
Chamber = 0

Agentt = ["Astra" , "Sova","Jett","Reyna","Chamber"]

for k in range(len(df["Agent"])):
    if df["Agent"].loc[k] == "Astra" :
      Astra = Astra + 1
    if df["Agent"].loc[k] == "Sova" :
      Sova = Sova + 1
    if df["Agent"].loc[k] == "Jett" :
      Jett = Jett + 1
    if df["Agent"].loc[k] ==  "Reyna" :
      Reyna = Reyna + 1
    if df["Agent"].loc[k] == "Chamber" :
      Chamber = Chamber + 1

agent = [Astra ,Sova,Jett,Reyna,Chamber]
color_2 = ["#9400D3","#008080","#D3D3D3","#8B008B","#DAA520"] 
plt.pie(agent , labels = Agentt ,colors=color_2, autopct='%.1f%%', startangle= 70)
title_text = {"size": 15 , "color":"#9932CC"}
plt.title("Agent",title_text)
plt.show()
print()
```

```python
exp = [0.2,0,0.2,0,0]
plt.pie(agent , labels = Agentt ,colors=color_2, autopct='%.1f%%',explode=exp, startangle= 70)
plt.title("Agent",title_text)
plt.show()
print()
```

### **Win rate (%)** 🥇

```python
win = 0
lose = 0

Resultss = ["Win","Lose"]

for z in range(len(df["Re_sult"])):
    if df["Re_sult"].loc[z] == "Win" :
      win = win + 1
    if df["Re_sult"].loc[z] == "Lose" :
      lose = lose + 1

plt.figure(figsize=[7,7])
color = ["#978A22","#767672"]
result = [win,lose]
plt.pie(result,labels = Resultss ,colors=color, autopct='%.1f%%')
plt.title(" Result ")
plt.show()
print()
```
